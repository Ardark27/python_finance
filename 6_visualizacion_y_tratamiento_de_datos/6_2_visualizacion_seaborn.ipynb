# Seaborn

- Seaborn es una librería enfocada a la representacion de datos estadísticos
- Tiene funciones más elaboradas que matplotlib
- Los conceptos básicos son iguales que en matplotlib

```python
import seaborn as sns
```

- Link al tutorial de [seaborn](https://seaborn.pydata.org/tutorial/introduction.html)
- Link a la galería de [ejemplos](https://seaborn.pydata.org/examples/index.html)

## Ejemplos útiles de seaborn

```python
penguins = sns.load_dataset("penguins") # Carga el dataset de los pinguinos que vamos a usar
```

```python
penguins.head() # Muestra las primeras 5 filas del dataset
```

### jointplot

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

```python
sns.jointplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde",
)
```

### pairplot

```python
sns.pairplot(data=penguins, hue="species")
```

### relplot

```python
sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g"
)
```

### violinplot

```python
tips = sns.load_dataset("tips")

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
               split=True, inner="quart", linewidth=1,
               palette={"Yes": "b", "No": ".85"})
sns.despine(left=True)
```

## Relación con matplotlib

- Podemos usar los figures y los axes en seaborn
- Podemos seguir usando todas las propiedades que habiamos de los gráficos de matplotlib

```python
import matplotlib.pyplot as plt

flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```

```python
f.savefig("heatmap.png")
```
